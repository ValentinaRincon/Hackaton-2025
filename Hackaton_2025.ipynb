Nuestra idea es implementar Barenstein Vazirani para poder encontrar el estado secreto del que no se acuerdan los desarrolladores. Para ello, iniciamos con todos los q-bits en |0> y aplicamos X en los bits que ya conocemos. Después, ponemos en superposición de estado el resto de q-bits para probar el resto de configuraciones, para posteriormente aplicar el black box (Uf), volver a aplicar Hadamard y medir. Según lo que se vió en clase, esto debería ser útil para hallar el string oculto. Sin embargo, al aplicar el protocolo no realizó lo deseado, no sabemos el porqué; creemos que nos pudo haber faltado un protocolo que no estamos teniendo presente.

```python

from condor_oracle import build_Uf_condor, N_CONFIG_BITS (importamos la el archivo de la black box)
Uf_condor = build_Uf_condor()
qc= QuantumCircuit(N_CONFIG_BITS+1)


indi_conocidos= [3,8,11]
for j in indi_conocidos:
    qc.x(j) # fijamos como 1 el estado de los qubits conocidos

indi_libres = [i for i in range(16) if i not in indi_conocidos]  
#aplicar hadamard a los qubits libres
for k in indi_libres:
    qc.h(k)

qc.x(16) #preparamos el qubit ancilla en |1>
qc.h(16) #preparamos el qubit ancilla en superposicion

qc.compose (Uf_condor, inplace=True)

for k in indi_libres:
    qc.h(k)

qc.measure_all()
#pintar circuito
qc.draw('mpl')


# No sabemos como imprimir el resultado del circuito sin 
```

```python
#Para sim
sim = AerSimulator()

qc_t= transpile(qc, sim)
resultado = sim.run(qc_t).result()
counts = resultado.get_counts()
print(counts)
```
